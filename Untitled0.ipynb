```python
import pandas as pd

final_df = pd.read_csv("merged_data_2010_2017.csv")
final_df.head()

```

```python
final_df.shape
final_df.columns
```

```python
final_df.info()
```

```python
rice_df = final_df[final_df['Crop'] == 'Rice']
wheat_df = final_df[final_df['Crop'] == 'Wheat']
```

```python
X_rice = rice_df[['Rainfall', 'Temperature']]
y_rice = rice_df['Yield']
```

```python
X_wheat = wheat_df[['Rainfall', 'Temperature']]
y_wheat = wheat_df['Yield']
```

```python
from sklearn.model_selection import train_test_split

Xr_train, Xr_test, yr_train, yr_test = train_test_split(
    X_rice, y_rice, test_size=0.2, random_state=42
)

Xw_train, Xw_test, yw_train, yw_test = train_test_split(
    X_wheat, y_wheat, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression

lr_rice = LinearRegression()
lr_wheat = LinearRegression()

lr_rice.fit(Xr_train, yr_train)
lr_wheat.fit(Xw_train, yw_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
```

```python
rice_pred = lr_rice.predict(Xr_test)

print("Rice RMSE:", np.sqrt(mean_squared_error(yr_test, rice_pred)))
print("Rice R²:", r2_score(yr_test, rice_pred))
```

```python
wheat_pred = lr_wheat.predict(Xw_test)

print("Wheat RMSE:", np.sqrt(mean_squared_error(yw_test, wheat_pred)))
print("Wheat R²:", r2_score(yw_test, wheat_pred))
```

```python
wheat_coef = pd.DataFrame({
    'Feature': X_wheat.columns,
    'Coefficient': lr_wheat.coef_
})
wheat_coef
```

```python
plt.scatter(yw_test, wheat_pred)
plt.xlabel("Actual Yield")
plt.ylabel("Predicted Yield")
plt.title("Wheat: Actual vs Predicted Yield")
plt.show()
```

```python
from sklearn.dummy import DummyRegressor
from sklearn.metrics import r2_score

dummy = DummyRegressor(strategy="mean")
dummy.fit(Xr_train, yr_train)
dummy_pred = dummy.predict(Xr_test)

print("Baseline R²:", r2_score(yr_test, dummy_pred))
print("Linear R²:", r2_score(yr_test, rice_pred))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(
    lr_rice, X_rice, y_rice, cv=5, scoring='r2'
)

print("Cross-val R² scores:", scores)
print("Mean CV R²:", scores.mean())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LinearRegression())
])

pipe.fit(Xr_train, yr_train)
pred = pipe.predict(Xr_test)

print("Scaled R²:", r2_score(yr_test, pred))
```
